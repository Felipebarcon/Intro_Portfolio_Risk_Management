```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in the csv file and parse dates
StockPrices = pd.read_csv('data/AAPL.csv', parse_dates=['Date'])
```

```python
StockPrices = StockPrices.sort_values(by='Date')
```

```python
StockPrices.set_index('Date', inplace=True)
```

```python
# Print only the first five rows of StockPrices
print(StockPrices.head(5))
```

```python
# Calculate the daily returns of the adjusted close price
StockPrices['Open'] = StockPrices['Close'].pct_change()
```

```python
# Check the first five rows of StockPrices
print(StockPrices.head(5))
```

```python
plt.hist(StockPrices["Returns"].dropna(), bins=75, density=False)
plt.show()
```

```python
# Plot the returns column over time
StockPrices['Returns'].plot()
plt.show()
```
